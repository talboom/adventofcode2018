```python
## PART 1
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# read file
file = open("input.txt","r")

# check file mode and readlines
if file.mode == 'r':
    data = file.read().splitlines()
```

```python
points = []
for line in data:
    x = int(line[10:16])
    y = int(line[17:24])
    u = int(line[36:38])
    v = int(line[39:42])
    points.append([x,y,u,v])
          
```

```python
size_min = [1e99,0] 
for s in range(10000,13000):
    width = max(x + s*u*1 for (x, y, u, v) in points) - min(x + s*u*1 for (x, y, u, v) in points)
    height = max(y + s*v*1 for (x, y, u, v) in points) - min(y + s*v*1 for (x, y, u, v) in points)
    size = width * height
    if size < size_min[0]:
        size_min = [size,s,width,height]
print(size_min)
```

```python
s = size_min[1]
max_w = max(x + s*u for (x, y, u, v) in points)
min_w = min(x + s*u for (x, y, u, v) in points)
max_h = max(y + s*v for (x, y, u, v) in points)
min_h = min(y + s*v for (x, y, u, v) in points)


textfield_x = []
textfield_y = []
for point in points:
    x,y,u,v = point
    textfield_x.append(x+s*u - min_w)
    textfield_y.append(-y-s*v + min_h)
```

```python
plt.figure(figsize=(size_min[2],size_min[3]))
plt.scatter(textfield_x,textfield_y)

```
